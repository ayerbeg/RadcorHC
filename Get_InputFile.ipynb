```python
#Packages
import numpy as np
import pandas as pd
```

```python
columns = ["Q2","W","g1","g1_stat","g1_sys","A1_F1","AF_stat","AF_sys"]

data = pd.read_csv("EG4_proton_g1data_4np.csv",skiprows=1,names=columns)

data["X"] = data.Q2/(data.W**2 - 0.938272**2 + data.Q2)
data.head(10)
```

```python
columns = ["Q2","W","g1","g1_stat","g1_sys","A1_F1","AF_stat","AF_sys"]

deu_data = pd.read_csv("g1D_Data.csv",skiprows=1,names=columns)

deu_data["xB"] = deu_data.Q2/(deu_data.W**2 - 0.938272**2 + deu_data.Q2)
deu_data.head(10)
```

```python
df = pd.concat([np.round(data["W"]**2,4),data["Q2"]],axis=1)

data.to_csv("Proton_Data.in", header=None, index=None, sep=",", mode="w")

df.to_csv('f1.in', header=None, index=None, sep=',', mode='w')
```

```python
deu_df = pd.concat([np.round(deu_data["W"]**2,4),deu_data["Q2"]],axis=1)

deu_data.to_csv("Deuteron_Data.in", header=None, index=None, sep=",", mode="w")

deu_df.to_csv('f1.in', header=None, index=None, sep=',', mode='w')
```
